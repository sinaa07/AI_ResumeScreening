```python
# Drop unnecessary columns
drop_col = ['Unnamed: 0', 'Job Description'] 
jobs = jobs.drop(columns=drop_col, errors='ignore')
```

```python
drop_col_res = ['ID', 'Resume','tokenized_resume'] 
resumes = resumes.drop(columns=drop_col_res, errors='ignore')
```

```python
res_text = resumes["extracted_skills"].fillna("") + " " + \
           resumes["experience_years"].fillna("").astype(str) + " " + \
           resumes["extracted_location"].fillna("")

jd_text = jobs["extracted_skills"].fillna("") + " " + \
          jobs["experience_required"].fillna("").astype(str) + " " + \
          jobs["extracted_location"].fillna("")
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
tfidf_vector = TfidfVectorizer()
matrix_res = tfidf_vector.fit_transform(res_text)
matrix_jd = tfidf_vector.transform(jd_text)
score = cosine_similarity(matrix_res, matrix_jd)
```

```python
def match(s):
    if s >= 0.6:
        return "Strong"
    elif s >= 0.3:
        return "Partial"
    else:
        return "No Match"
```

```python
def getMatch(id):
  selected_jd = jobs["jd_id"].iloc[id]
  jd_index = jobs[jobs["jd_id"] == selected_jd].index[0]

  similarities = cosine_similarity(matrix_res, matrix_jd[jd_index])[:, 0]

  matches2 = []
  for i, score in enumerate(similarities):
      if score >= 0.2:
          matches2.append([resumes["resume_id"].iloc[i], selected_jd, score, match(score)])

  matches2_df = pd.DataFrame(matches2, columns=["resume_id", "jd_id", "similarity_score", "match_category"])
  matches2_df = matches2_df.sort_values(by="similarity_score", ascending=False)
  return matches2_df
  
```

```python
df = getMatch(33)
df.to_csv("training_jd_33.csv", index=False)
df.head()
```

```python
import pandas as pd

def to_merged(matches_df, resumes_df, jobs_df, filename="training_data.csv"):
    resumes_filtered = resumes_df[["resume_id", "extracted_skills", "experience_years", "location"]].rename(
        columns={
            "extracted_skills": "resume_skills",
            "experience_years": "resume_experience",
            "location": "resume_location"
        }
    )

    jobs_filtered = jobs_df[["jd_id", "extracted_skills", "experience_required", "extracted_location"]].rename(
        columns={
            "extracted_skills": "jd_skills",
            "experience_required": "jd_experience",
            "extracted_location": "jd_location"
        }
    )
    training_data = matches_df.merge(resumes_filtered, on="resume_id", how="left") \
                              .merge(jobs_filtered, on="jd_id", how="left")

    training_data.to_csv(filename, index=False)
    print(training_data["match_category"].value_counts())
    print(f"Training data saved as {filename}!")
    return training_data
```

```python
to_merged(df,resumes,jobs,"training_jd_33.csv")
```
